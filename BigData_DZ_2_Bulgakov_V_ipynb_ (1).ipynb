Імпорт бібліотек

```python
import pandas as pd
import numpy as np
```

Створити DataFrame з 3 рандомних numpy маисвів.

```python
a1 = np.random.randint(1,9,10)
a2 = np.random.randint(1,9,10)
a3 = np.random.randint(1,9,10)
df = pd.DataFrame([a1,a2,a3])
df
```

Створити DataFrame з 5 Series які були створені на базі рандомних numpy маисвів розміру 12. Індекси першої Series повинні починатись з 0, другої з 1? третьої з 2 і так далі. Організувати DataFrame так, щоб дані були в стовпцях, а не рядках.

Кожен стовпець DataFrame - це Series.

```python
a1 = np.random.randint(1,9,10)
s1 = pd.Series(a1,index=np.arange(0,10))
a2 = np.random.randint(1,9,10)
s2 = pd.Series(a2,index=np.arange(1,11))
a3 = np.random.randint(1,9,10)
s3 = pd.Series(a3,index=np.arange(2,12))
a4 = np.random.randint(1,9,10)
s4 = pd.Series(a2,index=np.arange(3,13))
a5 = np.random.randint(1,9,10)
s5 = pd.Series(a2,index=np.arange(4,14))
df2 = pd.concat([s1,s2,s3,s4,s5],axis = 1)
df2
```

Переіменувати парні стовпці в отриманому DataFrame через функцію ".rename()".


```python
df2 = df2.rename(columns={0:"col0",2:"col2",4:"col4"})
df2
df2.columns = {"col0", "col1","col2","col3","col4"}
```

Перевірити чи є в цілій DataFrame NaN та перевірити в кожному окремому стовпці.


```python
F = df2.isnull().any().any()
print("NaN ", F)
F = df2.isnull().any()
print("NaN in Col: \n", F)
```

Замінити всі NaN в першому та третьому стовпці на медіану цього стовпця.


```python
F = df2["col0"].isnull()
F2 = df2["col2"].isnull()
data = df2.col0[~F].median()
print(data)
df2.col0 = df2.col0.replace(np.nan,data)
df2.col2 = df2.col0.fillna(data)
df2
```

Замінити всі NaN в другому та п'ятому стовпці на суму першого та останнього числа в цьому стовпці


```python
F = df2["col2"].isnull()
F1 = df2["col4"].isnull()
data = df2.col2[~F]
val = data[data.first_valid_index()] + data[data.last_valid_index()]
print(val)
df2.col2 = df2.col2.fillna(val)
df2.col4 = df2.col4.fillna(val)
df2
```

Замінити всі NaN в четвертому стовпці на число 1000.

```python
df2.col3 = df2.col3.fillna(1000)
df2
```

Знайти всі елементи більші за 3 та помножити їх на 2. Інші помножити на -1.


```python
F = df2 > 3
df2[F] *= 2
F1 = df2 < 3
df2[F1] *= -1
df2
```

В першому стовпці елементи, що менші за 5 піднести до кубу.





```python
Flags = df2 < 5
df2.col0[Flags.col0]  =df2.col0[Flags.col0]**3
df2
```

В останньому стовпці присвоїти значення 0 всім парним елементам.

```python
Flags = df2 % 2 == 0
df2.col4[Flags.col4] = 0
df2
```
